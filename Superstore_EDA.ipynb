```python
#importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#using pandas to import the data
pd.set_option("display.max_column",200)
s_data=pd.read_csv("SampleSuperstore.csv")
print(s_data)
```

```python
#display all columns
print(s_data.columns)
print(s_data.describe())
print(s_data.info())
```

```python
#Check columns with null values
print(s_data.isnull().sum()) #it is easier to compare the variables and plotting if there are no null values
```

```python
#Find the columns with Integer values
check1=s_data.select_dtypes(include=np.number)
print(check1)
check2=s_data.select_dtypes(include=object)
print(check2)
```

```python
#Drop duplicate values
dup1=s_data.drop_duplicates()
print(dup1)
```

```python
#Checking the region variable
region1=s_data['Region'].value_counts()
print(region1)
sns.set_style('whitegrid')
sns.countplot(s_data['Region'],order=s_data['Region'].value_counts().index)
plt.show()
```

```python
#Let's move to other columns
state1=s_data['State'].value_counts()
sns.set_style('darkgrid')
sns.countplot(s_data['State'],order=s_data['State'].value_counts().index)
plt.xticks(rotation=90)
plt.show()
```

```python
#Checking city status
city1=s_data['City']
print(city1.head())
```

```python
#Checking categories of product
sns.set_style("whitegrid")
sns.countplot(s_data['Category'],order=s_data['Category'].value_counts().index)
plt.show()
print(s_data['Category'].head())
```

```python
#Comparing some datasets
x=s_data['Category']
y=s_data['Region']
sns.set_style("whitegrid")
sns.displot(x="Region",y="Category",data=s_data)
plt.show()
```

```python
#Checking other sub category products
sns.set_style("whitegrid")
sns.countplot(s_data['Sub-Category'],order=s_data['Sub-Category'].value_counts().index)
plt.show()
print(s_data['Category'].head())
print(s_data.head())
```

```python
#checking sales as per cities vias
topcities=s_data.groupby('City')['Sales'].sum().sort_values(ascending=False).iloc[0:15]
plt.title('HIGHEST CITIES AS PER SALES')
plt.xlabel('CITY')
plt.ylabel('TOTAL SALES')
topcities.plot(kind='bar')
plt.show()

topprofits=s_data.groupby('Region')['Profit'].sum().sort_values(ascending=False).iloc[0:15]
sns.set_style("darkgrid")
plt.title('HIGHEST PROFITS AS PER SALES')
plt.xlabel('REGION')
plt.ylabel('TOTAL PROFITS')
topprofits.plot(kind='bar')
plt.show()
```

```python
#quantities as per sales
topqanatities=s_data.groupby('Quantity')['Sales'].sum().sort_values(ascending=False).iloc[0:15]
sns.color_palette("Greens")
plt.title('HIGHEST QUANTITIES AS PER SALES')
plt.xlabel('QUANTITY')
plt.ylabel('TOTAL SALES')
topcities.plot(kind='bar')
plt.show()
```

```python
#discount
topdiscount=s_data.groupby('City')['Discount'].sum().sort_values(ascending=False).iloc[0:15]
print(topdiscount)
```

```python
#comparing some columns for making overview
gernaral=s_data.groupby('State')['Profit','Sales'].sum()
sns.set_style("darkgrid")
gernaral.plot(kind='bar')
plt.show()

sns.set_style("whitegrid")
sns.countplot(s_data['Segment'])
```

```python
#checking out sales as per other products
plt.bar(x=s_data['Sub-Category'],height=s_data['Sales'])
sns.relplot(x=s_data['Sub-Category'],y=s_data['Sales'],data=s_data)
plt.xticks(rotation=90)
plt.show()

sns.swarmplot(x=s_data['Sales'],y=s_data['Country'],data=s_data)
plt.title('Sales acorrding to country')
plt.show()
```
